```python
#f = 'C:\\doit\\test'
#f = r'c:\doit\test'
f = 'subwaytime.csv'
```

```python
data = usecsv.switch(usecsv.opencsv(f))
```

```python
result = []

for row in data[2:]:
    result.append(row[10])
    
len(result)
```

```python
import matplotlib.pyplot as plt
result.sort()
plt.figure(figsize = (10, 5), dpi = 200)
plt.style.use('ggplot') #격자스타일
plt.bar(range(len(result)), result)
plt.show()
```

```python
mx = 0
mx_station = ''

for row in data[2:]:
    if sum(row[10:15:2]) > mx:
        mx = sum(row[10:15:2])
        mx_station = row[3] + ' ' + row[1]

print(mx_station, mx)
```

```python
mx = 0
mx_station = ''

t = int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요? '))

for row in data[2:]:
    a = row[4 + (t - 4) * 2]
    if a > mx:
        mx = a
        mx_station = row[3] + ' ' + row[1]

print(mx_station,mx)
```

```python
mx = 0
mx_station = ''

t = int(input('몇 시의 하차 인원이 가장 많은 역이 궁금하세요? '))

for row in data[2:]:
    a = row[5 + (t - 4) * 2]
    if a > mx:
        mx = a
        mx_station = row[3] + ' ' + row[1]

print(mx_station,mx)
```

```python
#가장 많이 승차하는 시간과 인원 수
mx = [0] * 24
mx_station = [''] * 24
for row in data[2:]:
    for j in range(24):
        a = row[j * 2 + 4]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3] + ' ' + row[1]
        # 시간 4, 5, 6, 7, 8 -> 24
        #index 4, 6, 8, 10, 12...
        # j    0, 1, 2, 3, 4
        # j * 2 + 4
        
for t, name in enumerate(mx_station):
    print(name, t + 4, '시', mx[t], '명')
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation = 90) #rotation 활용하여 x축 표기원활
plt.show()
```

```python
#하차인원
mx = [0] * 24
mx_station = [''] * 24
for row in data[2:]:
    for j in range(24):
        a = row[j * 2 + 5]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3] + ' ' + row[1]

        #index 5, 7, 9, 11...
        # j    0, 1, 2, 3, 4
        # j * 2 + 5
        
for t, name in enumerate(mx_station):
    print(name, t + 4, '시', mx[t], '명')
    
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation = 90) #rotation 활용하여 x축 표기원활
plt.show()
```

```python
#승하차인원
s_in = [0] * 24
s_out = [0] * 24

for row in data[2:]:
    for j in range(24):
        s_in[j] += row[j * 2 + 4] #승차인원
        s_out[j] += row[j * 2 + 5] #하차인원
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24), range(4, 28))
plt.show()
```
