#### np.arange()

```python
a = np.arange(-np.pi, np.pi, np.pi / 100)
plt.style.use('default')
plt.rcParams['axes.unicode_minus'] = False
plt.plot(a, np.sin(a), label = 'sin')
plt.plot(a, np.cos(a), label = 'cos')
plt.plot(a + np.pi / 2, np.sin(a), label = 'np.sin + 1.7')
plt.legend()
plt.show()
```

#### 버블차트 그리기

```python
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
#size = np.random.rand(100) * 100

mask1 = abs(x) > 50
mask2 = abs(y) > 50

x = x[mask1 + mask2] 
y = y[mask1 + mask2]

plt.scatter(x, y, c = x, cmap = 'jet', alpha = .7)
plt.colorbar()
plt.show()
```

```python
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
#size = np.random.rand(100) * 100

mask1 = abs(x) > 50
mask2 = abs(y) > 50

#x = x[mask1 + mask2] 
#y = y[mask1 + mask2]

plt.scatter(x, y, c = x, cmap = 'jet', alpha = .7)
plt.colorbar()
plt.show()
```

```python
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50
mask2 = abs(y) > 50

x = x[mask1 + mask2] 
y = y[mask1 + mask2]
size = np.random.rand(len(x)) * 100
print(len(x))

plt.scatter(x, y, s = size, c = x, cmap = 'jet', alpha = .7)
plt.colorbar()
plt.show()
```
