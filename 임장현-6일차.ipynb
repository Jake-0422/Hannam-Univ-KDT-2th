#### 인덱싱과 슬라이싱

#### 5번
문자열은 인덱싱을 지원한다. 양수로도, 음수로도 지원한다. 양수는 문자열의 첫 문자를 0부터 시작하여, 다음 문자로 갈 때마다 인덱스가 1씩 증가한다.
음수는 문자열의 마지막 문자를 -1부터 시작하여, 앞의 문자로 갈 때마다 인덱스가 1씩 감소한다.

```python
string = "PYTHON"

print(string[0])
print(string[1])
print(string[2])
print(string[5])

print(string[-1])
print(string[-3])
print(string[-6])
```

#### 7번
슬라이싱은 문자열의 특정한 부분을 추출하는 작업이다.

특정한 문자를 사용하기 위해선 "[]" 안에 하나의 인덱스를 넣은 반면, 특정한 부분을 사용하기 위해서는 []안에 인덱스의 범위를 넣는다. 예를 들어, [0:3]은 인덱스 0인 문자부터 인덱스 3인 문자 전까지 호출하는 것이다.

음수를 슬라이싱에 활용할 수도 있다. 이 경우, [작은 인덱스 : 큰 인덱스]의 형식으로 사용함에 유의해야 한다.

```python
string = "PYTHON"

print("*" * 5 + " 정방향 " + "*" * 5)
print(string[0 : 2])
print(string[4 : 5])

print("*" * 5 + " 역방향 " + "*" * 5)
print(string[-3 : -1])
print(string[-6 : -5])
```

#### P.48 신체 질량 지수를 파이썬으로 계산하기

자신의 키와 몸무게를 변수에 저장하고 이들 변수들을 이용하여 BMI를 계산해보자.
BMI는 신체 질량 지수로서 신장과 체중을 이용하여 계산할 수도 있다. BMI는 사람의 체지방량과 상관관계가 크다고 증명된 바 있다.
BMI 지수는 킬로그램 단위로 측정한 체중을 w, 미터 단위로 측정한 키가 h라고 할 때 다음과 같이 구할 수 있다.

> BMI = w/h*h

> 원하는 결과 = 체중(kg), 키(m)를 입력하면 자동으로 BMI 값이 계산되어 변수 bmi에 저장되고, 그 값을 다음과 같이 출력하도록 해보자.

bmi 24.1358024691358

```python
height = 1.80
weight = 78.2
bmi = weight / height**2
bmi
```
