```python
import pandas as pd
```

```python
P=[["A", 130, 1.9, 16.3],
    ["B", 250, 2.6, 10.2],
    ["C", 190, 2.2, 11.1],
    ["D", 300, 2.9, 7.1],
    ["F", 220, 2.3, 13.2],
    ["G",170, 2.2, 14.2]]

print(P)
```

```python
df = pd.DataFrame(P)
df
```

#### 1) 데이터 프레임 출력

```python
# columns 명칭 지정
df.columns=['Name','HP','Weight','Efficiency']
df
```

#### 2) name 열 인덱스 지정 출력

```python
#index를 Name으로 지정
df.set_index('Name',inplace=True)
df
```

#### 3) 마력과 연비를 곱한 값이 가장 큰 차종 출력

```python
df['HP*Efficiency']=df['HP']*df['Efficiency']
display(df)
```
