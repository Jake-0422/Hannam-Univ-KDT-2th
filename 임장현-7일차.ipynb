### 입력문 input()과 문자열함수

#### 1번
input()은 입력받은 정보를 반드시 stirng(문자열)의 형태로 전환한다. 따라서 아래 예시의 경우, a * b는 문자열끼리의 곱셈 연산을 시도한 것이기 때문에 에러가 발생한다.

```python
a = input()
b = input()
print(a * b)
```

이러한 에러를 방지하기 위해서는 input()을 통해서 입력받은 수를 정수타입으로 타입(type)을 변화시켜야 한다. 이 때 사용하는 함수가 int() 함수이다.

아래 예시에서는 int()를 활용하여 input()을 통해 입력받은 수를 정수형으로 형변환을 한다. 이 후, 변환된 정수형 변수를 저장하여 문자열끼리의 곱에 의한 에러를 방지한다.

```python
a = int(input())
b = int(input())
print(a * b)
```

### 4번
수를 4개 입력받아 평균과 분산을 계산해주는 프로그램을 작성하시오.

평균( y = x1 + x2 + x3 + x4 / 4)
분산 [(x1 - y) ** 2 + (x2 - y) ** 2 + (x3 - y) ** 2 + (x4 - y) ** 2] / 4

```python
a,b,c,d = input("네 수를 입력하세요:").split()

avg = (int(a)+int(b)+int(c)+int(d))/4
var = (((int(a)-avg)**2)+((int(b)-avg)**2)+((int(c)-avg)**2)+((int(d)-avg)**2))/4

print("평균:",avg,"분산:",var)
```

#### 2번
list의 내장 함수를 사용하여 numbers = [12.26, 5, 426.3, 0.221]의 항목의 개수, 최대값, 최솟값, 항목들의 합, 그리고 정렬된 리스트를 출력하는 프로그램을 작성하시오.

```python
numbers = [12.26, 5, 426.3, 0.221]

#항목의 개수
print(len(numbers))

#리스트 내 최대값
print(max(numbers))

#리스트 내 최솟값
print(min(numbers))

#항목들의 합
#print(sum(numbers)) > 계속 오류남

#항목 정렬
print(sorted(numbers))
```
