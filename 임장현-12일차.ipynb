### 7.8
사용자로부터 숫자 1보다 크고 10보다 작은 값 n을 입력으로 받아서 다음과 같이뱀의 몸통처럼 증가하는
이차원 배열을 출력하는 뱀행렬을 생성하여라. 이 행렬의 특징은 n이 5일때 5X5크기의 2차원 배열을 생성하는데,
이때 홀수 번째 행이 숫자가 증가하는 방향이 될 경우 짝수 번째 행은 숫자가 감소하는 형태를 가지는 특징이 있다.
이 문제를 반드시 for반복문과 리스트 슬라이싱을 사용하여 풀어보아라.

```python
pc=0   # print 된 횟수
ac=0   #append 된 횟수
n=int(input("  1~31중 하나의 숫자를 입력하세요. : "))
nl=[]  #숫자 리스트
print("\n  ",n,"에 대한 행렬은 다음과 같습니다.\n")
for i in range(1,n*n+1):   # n이 5일때 5*5까지의 숫자가 있어함. n이10일 경우 10*10까지의 숫자가 있어야함.
    if pc%2==0:            # 행을 나누는 기준은 pc가 짝수일 때로 정하였다. -> pc가 짝수 일 때 홀수 열이 됨.
        nl.append(i)        # 숫자리스트에 반복적으로 i를 append 합니다. append 후 ac를 한개씩 늘려줍니다.
        ac+=1
        if ac%n==0:   # 입력된 숫자 n만큼 행이 채워지면 다음 행으로 넘어가야합니다. 따라서 append된 횟수가 n이 될 때마다 print합니다.
            for j in nl[i-n:i+1]:  # 출력하고자 하는 만큼 리스트를 슬라이싱 해줍니다.
                if j<10:                 # 보기 좋게 나열하기 위한 조건문을 달아줍니다.
                    print(' ',j,end=' ')
                elif j<100:
                    print('',j,end=' ')
                else:
                    print(j,end=' ')
            pc+=1                   # 출력이 완료되면, pc를 하나 올려줍니다.
            print('')               #end로 인해서 다음 열이 옆에 붙는것을 막아줍니다.
    else:
        nl.append(i)
        ac+=1
        if ac%n==0:                 #리스트 형으로 출력하여도 되지만, 단위가 바뀔때마다 열이 안맞고, 문제지 답안에도 괄호가 없다!
                for j in nl[i:i-n-1:-1]:
                    if j<10:
                        print(' ',j,end=' ')
                    elif j<100:
                        print('',j,end=' ')
                    else:
                        print(j,end=' ')
                pc+=1
                print('')
```
